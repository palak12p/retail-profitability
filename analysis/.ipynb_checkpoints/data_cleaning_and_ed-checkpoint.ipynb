```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load data
df = pd.read_csv('../data/retail_sales_data.csv')
df['Date'] = pd.to_datetime(df['Date'])

# View basic info
df.head()
```

```python
df['Profit'] = df['Revenue'] - df['Cost']
```

```python
store_profit = df.groupby('Store')['Profit'].sum().sort_values(ascending=False)
print("üîù Top 5 Most Profitable Stores:")
print(store_profit.head())

store_profit.plot(kind='bar', figsize=(12,5), title='Store-wise Profit')
plt.ylabel("Profit (‚Çπ)")
plt.tight_layout()
plt.show()
```

```python
category_profit = df.groupby('Category')[['Revenue', 'Profit']].sum().sort_values(by='Profit', ascending=False)
print("\nüìä Category-wise Performance:")
print(category_profit)

category_profit['Profit'].plot(kind='bar', color='orange', title='Category-wise Profit', figsize=(8,4))
plt.ylabel("Profit (‚Çπ)")
plt.tight_layout()
plt.show()
```

```python
daily_revenue = df.groupby('Date')['Revenue'].sum()

plt.figure(figsize=(12,4))
plt.plot(daily_revenue)
plt.title("üìà Daily Revenue Trend")
plt.xlabel("Date")
plt.ylabel("Revenue (‚Çπ)")
plt.tight_layout()
plt.show()
```

```python
deadstock = df.groupby('Product')['Quantity'].sum().sort_values().head(10)
print("üõë Least Sold Products:")
print(deadstock)
```
